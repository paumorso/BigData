```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
```

```python
data_original = pd.read_csv('nyc-jobs.csv')
```

```python
data = data_original
```

```python
data.head(5)
```

```python
data.dtypes
```

```python
#Posting Date, Posting Updated, Process Date
data['Posting Date'] = pd.to_datetime(data['Posting Date'])
data['Posting Updated'] = pd.to_datetime(data['Posting Updated'])
data['Process Date'] = pd.to_datetime(data['Process Date'])
```

```python
data.head(5)
```

```python
data.dtypes
```

```python
#dataFiltrada = data[["Posting Date", "Job ID"]].transpose()

dataFiltrada = data[["Posting Date", "Job ID"]]

#pasar la fecha a indice para trabajarla
dataFiltrada.set_index(["Posting Date"], inplace=True)

GroupPostingDate = pd.DataFrame(dataFiltrada.groupby([(dataFiltrada.index.year),(dataFiltrada.index.month)]).count())
GroupPostingDate =  pd.DataFrame(GroupPostingDate["Job ID"].astype('int64', copy=False))

GroupPostingDate.plot()
```

```python
## dataFiltrada = data[["Posting Updated", "Job ID"]]
dataFiltrada = data[["Posting Updated", "Job ID"]]

#pasar la fecha a indice para trabajarla
dataFiltrada.set_index(["Posting Updated"], inplace=True)

GroupPostingUpdated = pd.DataFrame(dataFiltrada.groupby([(dataFiltrada.index.year),(dataFiltrada.index.month)]).count())
GroupPostingUpdated =  pd.DataFrame(GroupPostingUpdated["Job ID"].astype('int64', copy=False))

GroupPostingUpdated.plot()
```

```python
columnas = ["Agency", "Job ID"]
plot1 = data[columnas]
XPlot10 = plot1.groupby(["Agency"]).size().reset_index().rename(columns = {"Agency" : "Empleador", 0 : "Cantidad"})
XPlot10 = XPlot10.sort_values("Cantidad", ascending = False)
XPlot10 = XPlot10.head(40)
XPlot10.head(10)


XPlot10.plot.bar(x='Empleador', y='Cantidad', rot=60, color='green', figsize=(20,10))

plt.title('Top 10 de Empleadores', fontsize=18)
plt.xlabel('Empleador', fontsize=16)
plt.ylabel('Cantidad', fontsize=16)
plt.xticks(size = 8)
```

```python
#df = pd.DataFrame({'country': [177.0, 7.0, 4.0, 2.0, 2.0, 1.0, 1.0, 1.0]})
#df.index = ['USA', 'Canada', 'Russia', 'UK', 'Belgium', 'Mexico', 'Germany', 'Denmark']
#df = df.sort_values(by='country',ascending=False)
#df["cumpercentage"] = df["country"].cumsum()/df["country"].sum()*100

plt.rcParams['xtick.labelsize'] = 7

columnas = ["Agency", "Job ID"]
plot1 = data[columnas]

XPlot10 = plot1.groupby(["Agency"]).size().reset_index().rename(columns = {"Agency" : "Empleador", 0 : "Cantidad"})
XPlot10 = XPlot10.sort_values(by='Cantidad',ascending=False)
XPlot10 = XPlot10.head(40)
XPlot10["cumpercentage"] = XPlot10["Cantidad"].cumsum()/XPlot10["Cantidad"].sum()*100

fig, ax = plt.subplots(figsize=(20,10))
ax.bar(XPlot10["Empleador"], XPlot10["Cantidad"], color="C0")
for tick in ax.get_xticklabels():
    tick.set_rotation(60)

ax2 = ax.twinx()
ax2.plot(XPlot10["Empleador"], XPlot10["cumpercentage"], color="C1", marker="D", ms=7)
ax2.yaxis.set_major_formatter(PercentFormatter())

ax.tick_params(axis="y", colors="C0")
ax2.tick_params(axis="y", colors="C1")

plt.title('Pareto de los Top 10 Empleadores', fontsize=18)
#plt.xlabel('Empleador', fontsize=16)
#plt.ylabel('Cantidad', fontsize=16)


plt.show()

XPlot10

```
